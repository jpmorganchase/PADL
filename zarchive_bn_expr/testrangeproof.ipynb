```python
def compute_hash_modulo(b: int, order: int) -> int:
    # Step 1: Convert the integer `b` to bytes
    # Assuming big-endian byte order
    b_bytes = b.to_bytes((b.bit_length() + 7) // 8, byteorder='big')
    
    # Step 2: Concatenate `b` bytes with `0x00` as a single byte
    encoded = b_bytes + b'\x00'
    
    # Step 3: Compute SHA256 hash
    hash_digest = hashlib.sha256(encoded).digest()
    
    # Step 4: Convert hash to an integer
    hash_int = int.from_bytes(hash_digest, byteorder='big')
    
    # Step 5: Compute modulo with `order`
    result = hash_int % order
    return result

# Inputs
b = 532202063888764576930400187562339822033961532425119575628364409121098665955

# Compute the result
result = compute_hash_modulo(b, order)

print(f"Result: {result}")
```
